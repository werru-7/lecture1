```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Dokładny link RAW do Twojego pliku na GitHubie
url = 'https://raw.githubusercontent.com/werru-7/lecture1/main/Salary%20Data%20(2).csv'

# Wczytanie danych z GitHuba
# Używamy on_bad_lines='skip' w razie problemów z formatowaniem pliku
try:
    df = pd.read_csv(url)
    print("Dane wczytano pomyślnie!")
except Exception as e:
    print(f"Błąd przy wczytywaniu danych: {e}")
    print("Sprawdź, czy nazwa pliku na GitHubie to na pewno 'Salary Data (2).csv' i czy gałąź to 'main'.")

# Usunięcie ewentualnych wierszy z brakującymi danymi dla tych dwóch kolumn
df = df.dropna(subset=['Years of Experience', 'Salary'])

# Stworzenie wykresu punktowego (scatter plot)
plt.figure(figsize=(10, 6))
sns.scatterplot(data=df, x='Years of Experience', y='Salary', color='royalblue', s=60)

# Upiększanie wykresu (tytuły, osie)
plt.title('Zależność pensji od lat doświadczenia', fontsize=16)
plt.xlabel('Lata doświadczenia (Years of Experience)', fontsize=12)
plt.ylabel('Pensja (Salary w USD)', fontsize=12)
plt.grid(True, linestyle='--', alpha=0.7)

# Wyświetlenie wykresu
plt.show()
```
